```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
from aid_common_utils.file_utils import load_mask, load_dicom
import cc3d

statistics = torch.load('cl_nnunet_fold_stats.data', weights_only=False)
```

```python
keynames = ['name', 'shapes', 'spacing', 'is_abdominal_aorta', 'is_descending_aorta', 'is_arch_aorta', 'is_ascending_aorta']
filt_stats = {key: statistics[key] for key in keynames}
keynames = [
'dif_abdominal_aorta', 'dif_descending_aorta', 'dif_arch_aorta', 'dif_ascending_aorta', 'dif_binary',
'folds_abdominal_aorta', 'folds_descending_aorta', 'folds_arch_aorta', 'folds_ascending_aorta', 'folds_binary'
]
for key in keynames:
    filt_stats[f'mean_{key}'] = np.mean(statistics[key], axis=1)
    filt_stats[f'std_{key}'] = np.std(statistics[key], axis=1)
```

```python
df = pd.DataFrame(filt_stats)
df.head()
```

```python
files_for_checking = []
filt = df['is_arch_aorta'].values
keynames = ['abdominal_aorta', 'descending_aorta', 'arch_aorta', 'ascending_aorta']
upper_p, down_p = 90, 10
figsize = (12, 2)
fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"mean_dif_{key}"][filt], down_p)
    fig.add_subplot(1, 4, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'mean_dif_{key}'][filt], bins=50, color='limegreen')
    plt.axvline(x=thres, color = 'red')
    plt.xlabel('IoU')
    files_for_checking.extend(df['name'][filt][df[f"mean_dif_{key}"][filt] < thres].values.tolist())
fig.savefig('var1.png', dpi=fig.dpi)
plt.show()
```

```python
fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"std_dif_{key}"][filt], upper_p)
    fig.add_subplot(1, 4, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'std_dif_{key}'][filt], bins=50, color='hotpink')
    plt.axvline(x=thres, color = 'blue')
    plt.xlabel('IoU')
    files_for_checking.extend(df['name'][filt][df[f"std_dif_{key}"][filt] > thres].values.tolist())
fig.savefig('var2.png', dpi=fig.dpi)
plt.show()
```

```python
fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"mean_folds_{key}"][filt], down_p)
    fig.add_subplot(1, 5, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'mean_folds_{key}'][filt], bins=50, color='limegreen')
    plt.axvline(x=thres, color = 'red')
    plt.xlabel('IoU')
    files_for_checking.extend(df['name'][filt][df[f"mean_folds_{key}"][filt] < thres].values.tolist())
thres = np.percentile(df[f"mean_folds_binary"][filt], down_p)
fig.add_subplot(1, 5, 5)
plt.title(f'mean_folds_binary')
plt.hist(df[f'mean_folds_binary'][filt], bins=50, color='limegreen')
plt.axvline(x=thres, color = 'red')
plt.xlabel('IoU')
fig.savefig('var3.png', dpi=fig.dpi)
plt.show()
```

```python
fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"std_folds_{key}"][filt], upper_p)
    fig.add_subplot(1, 5, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'std_folds_{key}'][filt], bins=50, color='hotpink')
    plt.axvline(x=thres, color = 'blue')
    plt.xlabel('IoU')
    files_for_checking.extend(df['name'][filt][df[f"std_folds_{key}"][filt] > thres].values.tolist())

thres = np.percentile(df[f"std_folds_binary"][filt], upper_p)
fig.add_subplot(1, 5, 5)
plt.title(f'std_folds_binary')
plt.hist(df[f'std_folds_binary'][filt], bins=50, color='hotpink')
plt.axvline(x=thres, color = 'blue')
plt.xlabel('IoU')
files_for_checking.extend(df['name'][filt][df[f"std_folds_binary"][filt] > thres].values.tolist())
fig.savefig('var4.png', dpi=fig.dpi)
plt.show()
```

```python
files_for_checking = []
filt = df['is_arch_aorta'].values
keynames = ['abdominal_aorta', 'descending_aorta', 'arch_aorta', 'ascending_aorta']
upper_p, down_p = 90, 10
figsize = (12, 2)
fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
plt.suptitle('mean')
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"mean_dif_{key}"][filt], down_p)
    print(f'{key} = {thres:.03f} | ', end=' ')
    fig.add_subplot(1, 4, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'mean_dif_{key}'][filt], bins=50, color='limegreen')
    plt.axvline(x=thres, color = 'red')
    files_for_checking.extend(df['name'][filt][df[f"mean_dif_{key}"][filt] < thres].values.tolist())
plt.show()

fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
plt.suptitle('std')
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"std_dif_{key}"][filt], upper_p)
    print(f'{key} = {thres:.03f} | ', end=' ')
    fig.add_subplot(1, 4, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'std_dif_{key}'][filt], bins=50, color='hotpink')
    plt.axvline(x=thres, color = 'blue')
    files_for_checking.extend(df['name'][filt][df[f"std_dif_{key}"][filt] > thres].values.tolist())
plt.show()


fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
plt.suptitle('mean folds')
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"mean_folds_{key}"][filt], down_p)
    print(f'{key} = {thres:.03f} | ', end=' ')
    fig.add_subplot(1, 5, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'mean_folds_{key}'][filt], bins=50, color='limegreen')
    plt.axvline(x=thres, color = 'red')
    files_for_checking.extend(df['name'][filt][df[f"mean_folds_{key}"][filt] < thres].values.tolist())

thres = np.percentile(df[f"mean_folds_binary"][filt], down_p)
print(f'{key} = {thres:.03f} | ', end=' ')
fig.add_subplot(1, 5, 5)
plt.title(f'mean_folds_binary')
plt.hist(df[f'mean_folds_binary'][filt], bins=50, color='limegreen')
plt.axvline(x=thres, color = 'red')
files_for_checking.extend(df['name'][filt][df[f"mean_folds_binary"][filt] < thres].values.tolist())
plt.show()

fig = plt.figure(figsize=figsize)
fig.subplots_adjust(top=0.8)
plt.suptitle('std folds')
for idx, key in enumerate(keynames):
    thres = np.percentile(df[f"std_folds_{key}"][filt], upper_p)
    print(f'{key} = {thres:.03f} | ', end=' ')
    fig.add_subplot(1, 5, idx+1)
    plt.title(f'{key}')
    plt.hist(df[f'std_folds_{key}'][filt], bins=50, color='hotpink')
    plt.axvline(x=thres, color = 'blue')
    files_for_checking.extend(df['name'][filt][df[f"std_folds_{key}"][filt] > thres].values.tolist())

thres = np.percentile(df[f"std_folds_binary"][filt], upper_p)
print(f'{key} = {thres:.03f} | ', end=' ')
fig.add_subplot(1, 5, 5)
plt.title(f'std_folds_binary')
plt.hist(df[f'std_folds_binary'][filt], bins=50, color='hotpink')
plt.axvline(x=thres, color = 'blue')
files_for_checking.extend(df['name'][filt][df[f"std_folds_binary"][filt] > thres].values.tolist())
plt.show()
```
